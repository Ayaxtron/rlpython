```python
import gym
import numpy as np
```

```python
def epsilon_greedy_policy(Q, epsilon, actions):
    """ Q is a numpy array, epsilon between 0,1 
    and a list of actions"""
    
    def policy_fn(state):
        if np.random.rand()>epsilon:
            action = np.argmax(Q[state,:])
        else:
            action = np.random.choice(actions)
        return action
    return policy_fn

```

```python
env = gym.make('CliffWalking-v0')

# Initialize Q
Q = np.zeros((env.observation_space.n, env.action_space.n))

# Initialize M
M = {}
```

```python
# Here we go
n_episodes = 100
n_iter = 100

gamma = 1
alpha = 0.1

actions = range(env.action_space.n)

score = []    

for ep in range(n_episodes):
    done = False
    state = env.reset()
    policy = epsilon_greedy_policy(Q, epsilon=10./(ep+1), actions = actions)
    
    # Learning phase:
    while not done:
        action = policy(state)
        new_state, reward, done, _ = env.step(action)
        q_learn_step = reward + gamma*np.argmax(Q[new_state,:])-Q[state,action]
        Q[state, action] += alpha*(q_learn_step)
        M[(state,action)] = (new_state, reward)
        state = new_state

    # Planning phase:
    observed_states = list(set([x[0] for x in list(M.keys())]))
    actions_per_state = {s:list(set([a for a in actions \
                                     if (s,a) in list(M.keys())])) \
                         for s in observed_states}
    
    for _ in range(n_iter):
        state = np.random.choice(observed_states)
        action = np.random.choice(actions_per_state[state])
        new_state, reward = M[(state,action)]
        q_learn_step = reward + gamma*np.argmax(Q[new_state,:])-Q[state,action]
        Q[state, action] += alpha*q_learn_step
        state = new_state
```

```python
# Play with the learned policy
done = False
state = env.reset()
while not done:
        env.render()
        action = policy(state)
        new_state, reward, done, _ = env.step(action)
        state = new_state
```
