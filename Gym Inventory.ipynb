```python

def epsilon_greedy_policy(Q, epsilon, actions):
    """ Q is a numpy array, epsilon between 0,1 
    and a list of actions"""
    
    def policy_fn(state):
        if np.random.rand()>epsilon:
            action = np.argmax(Q[state,:])
        else:
            action = np.random.choice(actions)
        return action
    return policy_fn


env = InventoryEnv()
Q = np.zeros([env.observation_space.n, env.action_space.n])

gamma = 0.99 
alpha = 0.1
n_episodes = 10000


actions = range(env.action_space.n)

score = []    
done = False
state = env.reset()
policy = epsilon_greedy_policy(Q, epsilon=1./(j+1), actions = actions )

while not done:
    action = policy(state)
    new_state, reward, done, _ = env.step(action)
    new_action = policy(new_state)

    # Adjust to make it fit in [0,99]
    state -= 1
    action -= 1
    new_state -= 1
    new_action -=1

    Q[state, action] = Q[state,action] \
                + alpha*(reward+gamma*Q[new_state,new_action]-Q[state,action])

    state, action = new_state, new_action

    if len(score) < 100:
        score.append(reward)
        print("INFO: Running score: {}".format(np.mean(score)))
    else:
        score[j % 100] = reward

env.close()
```
